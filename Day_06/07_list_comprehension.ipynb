# List Comprehension

```python
nums = [3, 8, 10, 12]

squares = []
for num in nums:
    squares.append(num**2) # Square values then add it to list


print(squares)
```

```python
# List Comprehension

squares = [num**2 for num in nums]  # Pythonic way
# How it's read: Square all num for each num in nums
print(squares)
```

```python
odds = []
for i in range(50):
    if i % 2 != 0:
        odds.append(i)

odds = [num for num in range(50) if i % 2 != 0]

# Give num for each num in range(50) where it is odd

odds
```
