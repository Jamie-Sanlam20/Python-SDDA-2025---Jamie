# Ternary Operator

```py
result = value_if_true if condition else value_if_false
```

Ternary -> means 3 -> 3 operands:
- Statement 1
- Variable
- Statement 2



## Binary Operator (2 Operands)
- Arithmetic -> + , - , * , **, %
- Logical -> and, or, comparison operators (<, <=, >=, >, ==)

# Unary Operator (1 Operand)
- * - Unpacking
- not
- ~ (Bitwise operator)

````python
# Original

age = 10

if age >= 18:
    print("Eligible for driving test ✅")
else:
    print("Come after few years 👶🍼")
````

````python
age = 10

result = (
    "Eligible for driving test ✅" if age >= 18 else "Come after few years 👶🍼"
)  # Ternary operator
print(result)

# Statement 1 if condition is true else Statement 2
````
