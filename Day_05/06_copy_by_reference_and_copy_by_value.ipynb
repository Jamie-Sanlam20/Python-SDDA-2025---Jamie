# Copy by reference and Copy by value 

````python
prices = [1000, 1500, 400]
print("Original prices: ðŸ’°", prices)
prices1 = prices[:] # Copy by value
prices2 = prices # Copy by reference

prices.append(600)
prices1.append(800)
prices2.append(1000)

# What is output & Why?
print(prices) #  [1000, 1500, 400, 600, 1000]
print(prices1) # [1000, 1500, 400, 800]
print(prices2) # [1000, 1500, 400, 600, 1000]

# id -> memory address
print(id(prices))
print(id(prices1))
print(id(prices2))

print(id(prices) == id(prices2)) # They share same memory location
````

# How list are stored?
- Lists holds the base address of the first value (not the data)

```py
prices = [1000, 1500, 400]

# prices holds address of 1000 -> 1027
# prices[0] -> 1027 + 0
# prices[1] -> 1027 + 1
# prices[2] -> 1027 + 2

prices2 = prices  

# Both variable hold same address  -> Copy by reference
# price2 bases address also be 1027
```

Example information:
- copied values of prices to prices1 --> prices1 = prices[:] --> used split to copy the data into prices1
- prices1 has it's own base address; just the values were copied over.
- copied base address of prices to prices2 (any changes that's made to prices, happens to prices2 also)
