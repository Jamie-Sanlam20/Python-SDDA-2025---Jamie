## Task 1
On an average who (male or female) earns more tip each day

````python
tips_df.groupby(["day", "sex"]).aggregate({"tip": "mean"})

# or easier aggregate way:

# tips_df.groupby(["day", "sex"])['tip'].mean()
````

### Unstack
Transforms the result by pulling out "group by columns" into separate columns

- pulls out group by columns > making it easier to read

````python
# how to make it easier to read -> avoid nesting

# make female and male columns

# columns: day; female; male

tips_df.groupby(["day", "sex"]).aggregate({"tip": "mean"}).unstack()
````

## Task 2
On an average, max, min who (male or female) earns more tip each day
Clue: aggregate

````python
tips_df.groupby(['day', 'sex']).aggregate({'tip': ['mean', 'max', 'min']}).unstack()
````

## Task: Proportion of Meals
Create a pie chart comparing how many times people visited on which time of the day (Dinner, Lunch)
 

````python
y_no_of_people_per_time = tips_df.groupby('time')['time'].count()

y_no_of_people_per_time.plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Number of People Visiting Per Day Time'

)

# value count is a shorter way of group by
# data scientists people doesn't like pie charts - goves you relative percentages (not actual values)
# %% has a special meaning, so we have to say %% to get the normal percent
````

````python
plt.scatter(tips_df['total_bill'], tips_df['tip_percent'], color='plum')

plt.xlabel('Total Bill (R)')
plt.ylabel('Tip percent %')
plt.title('Tip percent vs Total Bill')
````
