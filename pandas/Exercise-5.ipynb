### Introduction:


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

users = pd.read_csv(url, sep='|')

users
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age = users.groupby('occupation')['age'].mean()

round(mean_age)
```

```python
avg_age = users.groupby('occupation')['age'].mean()

round(avg_age, 0)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_count = users[users['gender'] == 'M'].groupby('occupation')['gender'].count()

total_count = users.groupby('occupation')['gender'].count()

male_ratio = male_count / total_count

male_ratio.sort_values(ascending=False)
```

```python
male_count = users[users['gender'] == 'M'].groupby('occupation')['gender'].count()

total_count = users.groupby('occupation')['gender'].count()

male_ratio = male_count / total_count

male_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').aggregate({'age': ['min', 'max']})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean().unstack()
```

```python
users.groupby(['occupation', 'gender'])['age'].mean().unstack()
```

### Step 8.  For each occupation present the percentage of women and men

```python
(users.value_counts(['occupation', 'gender']) / 100).unstack()
```

```python
users_occupation = users.value_counts(['occupation', 'gender'])

users_occupation_percent = users_occupation / 100

users_occupation_percent.unstack()
```
