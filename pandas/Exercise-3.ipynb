```python
student_scores = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

student_scores_df = pd.DataFrame(student_scores)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(student_scores_df['preTestScore'], student_scores_df['postTestScore'])
```

```python
plt.scatter(student_scores_df['preTestScore'], student_scores_df['postTestScore'], s=student_scores_df['age'], color='red')
plt.xlabel('Pre Test Score')
plt.ylabel('Post Test Score')
plt.title('Pre vs Post Test Scores')
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
student_scores_df['sex'] = ['male', 'female', 'female', 'male', 'female']

color_map = {'male': 'blue', 'female': 'pink'}

colors = student_scores_df['sex'].map(color_map)

plt.scatter(student_scores_df['preTestScore'], student_scores_df['postTestScore'], s=student_scores_df['postTestScore']*4.5, color=colors)
```
