## DRY

D - Don't
R - Repeat 
Y - Yourself

- Logic should not be repeated (i.e., a + b)
- Put it in a container and re-use it as many times as you want (function)
- Logic repeated - Code Smell.

Why should code be DRY?
If the code isn't dry, if a bug arises, we'll have to fix it every place.
If not DRY - there is room for improvement (in code quality)

## Functions

Why should we use it?
- You only have to update the logic in one place.
- If a bug arises - we fix it at one place.
- It's easier to add features in one place.

How do we add a function?
Definition (called function declaration):

```py
def add(a, b):
    c = round(a + b, 2)
    return 
```

- add -> Function name

- Function body:
```py
c = round(a+b, 2)
return c
```
- *a, b* is called parameters.
- *1, 3* are arguments
- *parsing the argument* --> replaces a, b with 1 and 3
- *return* -> when return is called, the execution goes back to the line it was called (returning back to starting point)

How do we know that we calling a function?
the brackets:

```py
c1 = add(1, 3) # Execution starts
```

Function call:
```py
add(1,3)
```


-> 4 gets assigned to c1
-> when function is called, control transfers to definition, then `return` to starting point 



````python
def add(a, b):
    c = round(a + b, 2)

# because we didn't add a return, it's not assigning any value to c1

c1 = add(1, 3)
print(c1) # returns 'None'
````

-  Part 1 - c1 calls add function.
- Part 2 - add function executes and transfers it back to starting point
- Part 3 - assigns value to c1 (if there's no return, it doesn't assign it to variable - none gets parsed)

````python
def pattern(emo, num):
    for i in range(1, num +1):
        print(emo * i)

pattern("🥕", 5)
````

````python
books = [
    {
        "title": "Python Programming",
        "author": "Eric Matthes",
        "year": 2019,
        "available": True,
    },
    {
        "title": "Automate the Boring Stuff with Python",
        "author": "Al Sweigart",
        "year": 2020,
        "available": True,
    },
    {
        "title": "Learning Python I",
        "author": "Mark Lutz",
        "year": 2013,
        "available": False,
    },
    {
        "title": "Fluent Python",
        "author": "Luciano Ramalho",
        "year": 2015,
        "available": True,
    },
    {
        "title": "Adavance Python",
        "author": "Mark Lutz",
        "year": 2015,
        "available": False,
    },
]

def new_book():
      title = input("Please tell me the title: ")
      author = input("Please tell me the author: ")
      year = input("Please tell me the year: ")
      available = input("Please tell me the availability: ")

      new_book = {"title": title, "author": author, "year": year, "available": available}
      books.append(new_book)
      print("Successfully added 😄✅")

def all_books():
    print(books)

def close():
    print("Bye 😄👋")
    return 'n'

def main_menu(menu):
    if menu == 1:
        new_book()
    elif menu == 2:
        all_books()
    elif menu == 3:
        return close()
    else:
        print("Sorry, selection isn't available")
    return 'y'

print("Welcome to our Library app 📚")

main = 'y'

while main == 'y':
    print("""
Main menu:
1. Add book to the library
2. Print all the books
3. Exit""")

    menu = int(input("Please choose an option: "))
    main = main_menu(menu)
````
